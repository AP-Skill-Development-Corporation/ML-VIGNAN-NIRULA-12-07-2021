### Data Preprocessing using scikit-Learn

- Problems:
    - insufficient of data
    - outliers
    - missing data
    - too much of data
    - duplicate data
- Scaling techniques
    - Standarad scalar
    - Robust scalar
    - Mix-Max scalar/range scalar

```python
from sklearn import preprocessing
```

```python
print(dir(preprocessing))
```

```python
print(help(preprocessing.minmax_scale))
```

```python
data.head()
```

```python
data1 = data[["Sales","Discount","Order_Quantity","Profit","Shipping_Cost","Product_Base_Margin"]]
```

```python
data1.head()
```

```python
sc = preprocessing.StandardScaler()
```

```python
sc_t = sc.fit_transform(data1)
```

```python
sc_t
```

```python
data1.columns
```

```python
sc1_data = pd.DataFrame(sc_t,columns = data1.columns)
sc1_data.head()
```

#### Robust scaler

```python
rs = preprocessing.RobustScaler()
```

```python
rs_t = rs.fit_transform(data1)
```

```python
rs_t
```

```python
pd.DataFrame(rs_t,columns=data1.columns).head()
```
