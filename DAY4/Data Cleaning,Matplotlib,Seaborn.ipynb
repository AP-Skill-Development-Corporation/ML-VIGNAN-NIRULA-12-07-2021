### Data Cleaning:

- Missing data [repalce,fillna,dropna]
- fill missing values using sklearn
- duplicate data

```python
import pandas as pd
import numpy as np
```

```python
a = np.array([[1,2,np.nan,3,4],[10,22,34,67,89],[23,45,89,67,90],
          [np.nan,45,90,np.nan,90],[23,np.nan,90,67,89]])
```

```python
a
```

```python
a.shape
```

```python
d = pd.DataFrame(a,columns = ["one","two","three","four","five"],index=["a","b","c","d","e"])
d
```

```python
d.columns
```

```python
d.index
```

```python
d.isnull()
```

```python
d.isnull().sum()
```

```python
# we can handle null values in two ways
# dropna
# fillna
```

```python
d.dropna()
```

```python
d
```

```python
d.dropna(axis=1)
```

```python
d.replace(np.nan,0)
```

```python
d
```

```python
d["one"] = d["one"].replace(np.nan,0)
```

```python
d
```

```python
d["two"].mean()
```

```python
d["two"].replace(np.nan,d["two"].mean())
```

```python
d["two"] = d["two"].fillna(d["two"].mean())
```

```python
d["three"].median()
```

```python
d["three"] = d["three"].fillna(d["three"].median())
```

```python
d
```

```python
s = pd.DataFrame({"lan":["english",np.nan,"telugu","english"],"alp":["a",np.nan,"b",np.nan]})
```

```python
s
```

```python
s["lan"].fillna(method="bfill")
```

```python
s["alp"].fillna(method="ffill")
```

### Piechart

```python
a = ["python","ml","da","ai"]
b = [70,45,60,15]
c = ["pink","gold","yellow","skyblue"]
plt.pie(b,labels=a,autopct = '%1.3f%%',colors = c)
plt.show()
```

```python
a = ["python","ml","da","ai"]
b = [70,45,60,15]
c = ["pink","gold","yellow","skyblue"]
e = (0,1,1,1)
plt.pie(b,labels=a,autopct = '%1.3f%%',colors = c,explode=e)
plt.show()
```
